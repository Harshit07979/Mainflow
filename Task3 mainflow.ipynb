# CRATING BAR PLOT


```python
import matplotlib.pyplot as plt

def create_bar_charts(data, columns):
    for col in columns:
        if col not in data.columns:
            print(f"Error: Column '{col}' does not exist in the DataFrame.")
            continue
        
        if data[col].isnull().any():
            print(f"Warning: Column '{col}' contains missing values.")
        
        years = data['year']
        values = data[col]
        
        plt.figure(figsize=(10, 5))
        plt.bar(years, values, color='blue')
        plt.xlabel('Year')
        plt.ylabel(col.replace('_', ' ').title())
        plt.title(f'{col.replace("_", " ").title()} by Year')
        plt.grid(True)
        plt.show()


```

```python
columns_to_plot = [ 'tot_hhs', 'own', 'own_wm', 'own_prop', 'own_wm_prop',
       'prop_hhs', 'age', 'size', 'income', 'expenditure', 'eqv_income',
       'eqv_exp']
create_bar_charts(data, columns_to_plot)
```

## CREATING LINE PLOT

```python
def create_line_plot(data,columns):
    for col in columns:
        if col not in data.columns:
            print(f"Error: Column {col} is not found in data")
            continue
            
        if data[col].isnull().any():
            print(f"Error:Column {col} contains null value")
            continue
            
        years = data['year']
        values = data[col]
        
        plt.figure(figsize=(10,6))
        plt.plot(years,values,marker='o', linestyle='-',color='green')
        plt.xlabel('year')
        plt.ylabel(col.replace("_"," ").title())
        plt.title(f'{col.replace("_", " ").title()} by Year')
        plt.legend([col.replace('_', ' ').title()])
        plt.grid(True)
        plt.show()
```

```python
columns_to_plot = [ 'tot_hhs', 'own', 'own_wm', 'own_prop', 'own_wm_prop',
       'prop_hhs', 'age', 'size', 'income', 'expenditure', 'eqv_income',
       'eqv_exp']
create_line_plot(data, columns_to_plot)
```
